**Задание 1 (2 балла).**

Отчистите текст от мусора (тэгов, хешей и тп) с помощью регулярных выражений. Постарайтесь убрать весь мусор, но если что-то удалить не получается, то не мучайтесь. Главное, чтобы мусор не проявлялся в результатах следующих заданий.

```python
with open('zhivago.txt') as f:
    text = f.read()
    
print(text[:2000])
```

```python
import re

text = re.sub('<lang>([^<>]*)</lang>', ' ', text)
text = re.sub('<nickname>([^<>]*)</nickname>', ' ', text)
text = re.sub('<program-used>([^<>]*)</program-used>', ' ', text)
text = re.sub('<date>([^<>]*)</date>', ' ', text)
text = re.sub('<src-url>([^<>]*)</src-url>', ' ', text)
text = re.sub('<id>([^<>]*)</id>', ' ', text)
text = re.sub('<version>([^<>]*)</version>', ' ', text)
text = re.sub('<binary.*?>([^<>]*)</binary>', ' ', text)
text = re.sub(u'\xa0', ' ', text)

print(text[:2000])
```

```python
def remove_html_tags(text):
    text = re.sub(r'<[^>]+>', ' ', text)
    text = re.sub(r'\n', ' ', text)
    text = re.sub('  +', ' ', text)
    return text

text = remove_html_tags(text)
print(text[:2000])
```

**Задание 3. (2 балла)**

Сделайте стемминг и найдите по несколько частотных примеров на каждый из видов ошибок:

1) два разных слова ошибочно свелись к одинаковой основе

2) слово не изменилось после стемминга (слово должно быть русским и длиннее 4 символов)

***не делайте это задание полностью вручную, напишите правила, которые отловят потенциальные ошибки и выберите из них***


```python
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import word_tokenize, wordpunct_tokenize

stemmer = SnowballStemmer('russian')

stemm_list_1 = list()
stemm_list_2 = list()
for sentence in token_sentences:
    for token in sentence:
        stemm = stemmer.stem(token)
        if token != stemm:
            stemm_list_1.append(stemm)
        stemm_list_2.append((token, stemm))
        
Counter(stemm_list_1).most_common(20)
```

```python
task_1 = dict()
task_1.update({'как': set()})
task_1.update({'не': set()})
task_1.update({'друг': set()})

for sentence in token_sentences:
    for token in sentence:
        stemm = stemmer.stem(token)
        if stemm in task_1:
            task_1[stemm].add(token)
print(task_1)
```

**Ответ 1:** Стеммер приводит частицу "как" и различные словоформы леммы "какой" к одной основе "как". А частицу "не" и различные словоформы местоимения "ней" к одной основе "не". Тоже самое с существительным "друг" и словоформами леммы "другой".

```python
stemm_list_2[:5]
```

```python
new_list = list()
for pair in stemm_list_2:
    if pair[0] == pair[1] and len(pair[0]) > 4:
        new_list.append(pair[0])
        
Counter(new_list).most_common(5)
```
